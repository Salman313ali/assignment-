```python
import pandas as pd
data = pd.read_csv("bbc-text.csv")
```

```python
data
```

```python
all_text = data["text"].values
all_category = data["category"].values
```

```python
def into_array(txt):
    from sklearn.feature_extraction.text import CountVectorizer
    vectorizer = CountVectorizer()
    vectorizer.fit(txt)
    vector = vectorizer.transform(txt)
    vector = vector.toarray()
    return vector
```

```python
vector_text = into_array(all_text)
vector_category = into_array(all_category)
```

```python
train_text = vector_text[:2000]
test_text = vector_text[2000:]
train_category = vector_category[:2000]
test_category = vector_category[2000:]
```

```python
x_val = train_text[:500]
partial_x_train = train_text[500:]
y_val = train_category[:500]
partial_y_train = train_category[500:]
```

```python
from keras import models
from keras import layers
model = models.Sequential()
model.add(layers.Dense(500,activation = 'relu', input_shape = (train_text.shape[1],)))
model.add(layers.Dense(200,activation = 'relu'))
model.add(layers.Dense(100,activation = 'relu'))
model.add(layers.Dense(5,activation = 'softmax'))
```

```python
model.compile(optimizer='Adam',
loss='categorical_crossentropy',
metrics=['acc'])
history = model.fit(partial_x_train,partial_y_train,epochs=20,batch_size=128,validation_data=(x_val, y_val))
```

```python
result = model.evaluate(test_text, test_category)
```

```python
print("acccuracy = ",result[1]*100,"%")
print("loss = ",result[0]*100)
```
